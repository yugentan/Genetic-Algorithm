# Hyper-parameter exploration

<B>Abstract</B>
<Br>
The purpose of this study is to explore the difference in parameter setting of the genetic algoritm in relation to the evolution proccess. Performace of the algorithm are experimented and observed through the different configuration of parameters, mutation rate and population. Multiple runs are conducted for each parameter configuration and the speed of the evolution is measured by the increase of the population's mean fitness over time (generation count). The findings of each configuration are tabulated into csv files(which i will provide a github link to) and graphed with pandas and matplotlib to visualise the influence each parameter have over the evolution speed. Conclusion drawn from the runs should determine an optimal approach to the configuration for achieving faster evolution rate of the genetic algorithm.
   

<B> Experiments </B>
<Br>
<B><U>Experiment Setup (Mutation)</U></B>
<Br>
    For the experiment on mutation rate, the mean fitness (<B>distance travelled</B>) will be measured against the generation count (<B>time taken</B>). The type of mutation rate of <B>point mutate/shrink mutate/grow mutate</B>, will be measured from the range of <B>0.01 to 0.25</B>. In any given case where the type of mutation is <B>not in scope</B> of the experiment, the <B>default</B> mutation rate will be set to <B>0.25</B>.For the fairness of the experiment, a population size of <B>10</B> and a gene count of <B>3</B> will be used and the program will be running on <B>4</B> threads. Measurement of the fitness will be taken untill the population of genome have evolved over <B>100</B> generations. 

| Mutation        | Mutation Rate | Gene count | Population Size | Thread Used | Generation |
|-----------------|---------------|------------|-----------------|-------------|------------|
| Point Mutate    | 0.01 - 0.25   | 3          | 10              | 4           | 100        |
| Shrink Mutate   | 0.01 - 0.25   | 3          | 10              | 4           | 100        |
| Grow Mutate     | 0.01 - 0.25   | 3          | 10              | 4           | 100        |



<B><U>Experiment Setup (Population)</U></B>
<Br>
    For the experiment on population size, the mean fitness (<B>distance travelled</B>) will be measured against the generation count (<B>time taken</B>). The population size used will be taken from the range of <B>2 - 10</B> with the <B>default</B> mutation rate of <B>0.25</B> across all mutation and a gene count of <B>3</B> and the program will run on <B>4</B> threads. The experiment will end when the population of genome have evolved over <B>100</B> generations.

<B>Evaluation Results</B>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files=["pm"+ str(i+1) +".csv" for i in range(25)]
dfs=[]
for file in files:
    df = pd.read_csv("../mutationdata/point_mutation/"+file,
                     header=None)
    dfs.append(df)
```

```python
plt.figure(figsize=(10, 8))

for i, df in enumerate(dfs):
    plt.plot(df.iloc[:, 0], df.iloc[:, 1], 
             label=f"Point Mutation Rate = {float(i+1)/100:.2f}")

plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.xlabel("Generation Count")
plt.ylabel("Mean Fitness")
plt.title("Point Mutation Time Series Chart")
plt.show()
plt.show()
```

# Experiment with encoding scheme

In this part of the experiment, we will conduct 3 different augmentation in the scale of the genome class get_gene_spec() function. We will run <B>100</B> generations with a population size of <B>10</B>, a gene count of <B>3</B>, with default mutation rate of <B>0.25</B> on <B>4</B> threads.
<Br>

<B><U>Augmenting Link-Length </U></B>
</Br>
We will keep the rest of the scale of the genome class at default. The scale setting of link-length will be tuned to <B>5</B>. After the 100 generation, we get the follow image as the resultant creature.

Base on creature produce, we can tell that the link-length property in the encoding scheme affects the length of each link. This means that the higher the scale you were to assign link-length in the genome class get_gene_spec() function, you will get a creature with link that are of more cylindrical shape.

<B><U>Augmenting Link-Radius </U></B>
<Br>
The next property i have experimented with is the link radius. I will keep the link length the same as before so that i can get a better result as to if the link radius affect the thin/thick-ness of the creature.The link radius scale of this experiment is set to <B>0.5</B>. This generation was paused on generation 41. As you can see in the video provided as well, this particular generation had a spike in its mean fitness around 20th cycle. This could only be the result of either the creature has learn to fall below ground level and can accumalate very high mean fitness/distance covered or it has learnt to fling itself and fly in the environment to cover a large distance. The image below depicts the creature on the 41st generation.

Base on the creature produce, we can see that it definitely have signs of links being thinner as compared to the previous creature. Thus, this means that the higher scale the link radius is, the thicker the link is and the lower scale the link radius is, the thinner the link is.
